```python
def sql_query(query):
    result = engine.execute(query)
    df = []
    [df.append(row) for row in result]
    df = pd.DataFrame(df)
    return df
```

```python
query2 = text('''SELECT *, DATE_FORMAT(rental_date, "%W") as day,
CASE
WHEN DATE_FORMAT(rental_date, "%W") = 'Saturday' or 'Sunday' then 'weekend'
ELSE 'workday'
END AS 'day_type'
FROM rental;''')
```

<h5>Query 6: Which kind of movies (rating) have a mean duration of more than two hours?<h5>

```python
query6 = '''
SELECT title, rating, length from film
WHERE (length) > 120
GROUP BY title, rating, length;
'''
```

```python
query6_result= sql_query(query6)
query6_result
```
